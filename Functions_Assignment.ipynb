5. What are iterators in Python and how do they differ from Iterables?

 Ans: In Python, iterables and iterators are fundamental concepts for working with sequences of data. They provide a powerful and efficient way to process collections of elements one at a time.

 Iterables: An iterable is any object that can be looped over using a for loop.

 The key feature is that it has an __iter__() method that returns an iterator object. Examples: lists, tuples, strings, dictionaries, sets, and custom objects that implement the __iter__() method.

 Iterators: An iterator is an object that represents a stream of data. It remembers its current position and provides access to the next element in the sequence. It has a __next__() method that returns the next element in the sequence. It raises a StopIteration exception when there are no more elements. Every iterator is also an iterable (because it has the __iter__() method, which typically returns itself).

 The key difference is that iterable can be looped over whereas iterator provides access to elements one at a time. Iterators doesn't necessarily remember its position while iterables remembers its current position

6. Explain the concept of generators in Python and how they are defined.

 Ans:  Generators are a simple way to create iterators using functions and the yield keyword instead of returning values

 Key Characteristics of Generators:

 ·        Use the yield keyword: Instead of using the return statement, generators use the yield keyword. When yield is encountered, the generator function pauses its execution, saves its current state, and returns the value yielded. The next time the generator is called, it resumes execution from where it left off.

 ·        Create iterators: Generators automatically create an iterator object   when called.

 ·        Memory Efficient: Since generators produce values on-demand, they consume much less memory compared to creating a full list of values upfront.

 ·        Useful for Infinite Sequences: Generators can be used to generate infinite sequences, such as Fibonacci numbers or prime numbers

 Generators are a powerful feature in Python that can significantly improve the performance and efficiency of your programs, especially when dealing with large or infinite sequences.

7. What are the advantages of using generators over regular functions?

 Ans: Generators in Python offer several key advantages over regular functions:

 i) Memory Efficiency:

 ·        Generators produce values on-the-fly, one at a time.

 ·        This means they don't need to store the entire sequence in memory, making them ideal for working with large datasets or infinite sequences.

 ·        This significantly reduces memory consumption compared to regular functions that would create and store the entire sequence before returning it.

 ii) Laziness (On-demand Evaluation):

 ·        Generators only calculate values when they are explicitly requested (e.g., using next() or in a loop).

 ·        This "lazy" evaluation can improve performance, especially for computationally expensive operations, as unnecessary calculations are avoided.

 iii) Conciseness and Readability:

 ·        Generators often lead to more concise and readable code, especially for iterative tasks.

 ·        The yield keyword provides a clean way to express the generation of a sequence without explicitly constructing intermediate data structures.

 iv) State Preservation:

 ·        Generators automatically maintain their internal state between calls to next().

 ·        This simplifies the management of state variables compared to manually tracking them in regular functions.

 v) Suitable for Infinite Sequences:

 ·        Generators can easily represent infinite sequences, such as Fibonacci numbers or prime numbers, as they don't need to store the entire sequence in memory.

 In summary, generators provide a powerful and efficient mechanism for working with sequences of data in Python, offering significant advantages in terms of memory usage, performance, and code readability.

```python
# Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_of_evens(numbers):
    """
    Returns the sum of all even numbers in a list.

    Args:
        numbers (list): A list of integers.

    Returns:
        int: The sum of all even numbers in the list.
    """
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum


numbers = [1, 2, 3, 4, 5, 6]
result = sum_of_evens(numbers)
print(f"The sum of even numbers is: {result}")
```

```python
# Create a Python function that accepts a string and returns the reverse of that string.
def reverse_string(s):
    """
    Returns the reverse of the input string.

    Args:
        s (str): The string to reverse.

    Returns:
        str: The reversed string.
    """
    return s[::-1]

input_string = "hello"
reversed_string = reverse_string(input_string)
print(f"The reverse of '{input_string}' is '{reversed_string}'")
```

```python
#  Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.
def square_numbers(numbers):
    """
    Returns a list of the squares of each number in the input list.

    Args:
        numbers (list): A list of integers.

    Returns:
        list: A new list containing the squares of each number.
    """
    return [num ** 2 for num in numbers]

input_list = [1, 2, 3, 4, 5]
squared_list = square_numbers(input_list)
print(f"The squares of {input_list} are {squared_list}")
```

```python
#  Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    """
    Checks if a given number is prime.

    Args:
        n (int): The number to check.

    Returns:
        bool: True if the number is prime, False otherwise.
    """
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

prime_numbers = [num for num in range(1, 201) if is_prime(num)]

print(f"Prime numbers from 1 to 200: {prime_numbers}")
```

```python
#  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:
    """
    An iterator class to generate the Fibonacci sequence up to a specified number of terms.
    """
    def __init__(self, num_terms):
        """
        Initialize the iterator with the number of terms to generate.

        Args:
            num_terms (int): The number of terms in the Fibonacci sequence to generate.
        """
        self.num_terms = num_terms
        self.current = 0
        self.next = 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.num_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return self.current
        elif self.count == 1:
            self.count += 1
            return self.next
        else:
            fib = self.current + self.next
            self.current, self.next = self.next, fib
            self.count += 1
            return fib
num_terms = 10
fib_iterator = FibonacciIterator(num_terms)

print(f"Fibonacci sequence with {num_terms} terms:")
for term in fib_iterator:
    print(term, end=" ")
```

```python
# Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(max_exponent):
    """
    Generator function that yields powers of 2 up to the given exponent.

    Args:
        max_exponent (int): The maximum exponent to compute 2^n.

    Yields:
        int: The powers of 2 from 2^0 to 2^max_exponent.
    """
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent

max_exponent = 5
print(f"Powers of 2 up to 2^{max_exponent}:")
for power in powers_of_two(max_exponent):
    print(power, end=" ")
```

```python
#  Implement a generator function that reads a file line by line and yields each line as a string.
example_file_path = "example.txt"
with open(example_file_path, 'w') as file:
    file.write("Line 1: This is the first line.\n")
    file.write("Line 2: Here is the second line.\n")
    file.write("Line 3: And this is the third line.\n")


def read_file_line_by_line(file_path):
    """
    Generator function that reads a file line by line and yields each line as a string.

    Args:
        file_path (str): Path to the file to read.

    Yields:
        str: The next line in the file.
    """
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()

print(f"Reading lines from '{example_file_path}':")
for line in read_file_line_by_line(example_file_path):
    print(line)
```

```python
#  Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
celsius_temperatures = [0, 10, 20, 30, 40]
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))
print("Celsius temperatures:", celsius_temperatures)
print("Fahrenheit temperatures:", fahrenheit_temperatures)
```

```python
# Create a Python program that uses `filter()` to remove all the vowels from a given string.
def is_not_vowel(char):
    vowels = "aeiouAEIOU"
    return char not in vowels
input_string = "Hello, World!"
filtered_string = ''.join(filter(is_not_vowel, input_string))
print("Original string:", input_string)
print("String without vowels:", filtered_string)
```
